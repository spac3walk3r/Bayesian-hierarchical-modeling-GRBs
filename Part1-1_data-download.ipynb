```python
def GRB_preparation(GRB_name):
    
    
    download_info = download_GBM_trigger_data(grb_info['trigger'], detectors=grb_info['detectors'])

    det_b_erange = ('250-30000')
    det_n_erange = ('8.1-900')

    bak_pre = (grb_info.get("background")).get("pre")
    bak_post = (grb_info.get("background")).get("post")

    bak_pre_start, bak_pre_stop = re.findall(r"-\d*\.\d+", bak_pre)
    bak_pre_start, bak_pre_stop = float(bak_pre_start), float(bak_pre_stop)
    bak_post_start, bak_post_stop = re.findall(r"\d*\.\d+", bak_post)
    bak_post_start, bak_post_stop = float(bak_post_start), float(bak_post_stop)


    fluence_interval = (grb_info.get('source')).get('fluence')
    fluence_start, fluence_stop = re.findall(r"\d*\.\d+", fluence_interval)
    fluence_start, fluence_stop = float(fluence_start), float(fluence_stop)
    if(fluence_interval[0]=='-'):
        fluence_start = fluence_start*(-1)   
    
    
    # creating TS objects
    
    dets_ts = {}
    dets_spectra = {}

    for det in dets:
        det_ts = TimeSeriesBuilder.from_gbm_tte(det,
                                                tte_file = download_info[det]['tte'],
                                                rsp_file = download_info[det]['rsp'],unbinned=True)
        dets_ts[det] = det_ts

    for det in dets:
        dets_ts[det].set_background_interval(bak_pre,bak_post)

                
    # saving backgrounds as ht5 files

    temp_dir = 'backgrounds'
    if(path_exists_and_is_directory(temp_dir)==False):
        if_directory_not_existing_then_make(temp_dir)
        with within_directory(temp_dir):
            for det in dets:
                dets_ts[det].save_background('%s_%s'%(GRB_name,det), overwrite=True)

          
    # setting active time intervals and selecting en. ranges
                
    for det in dets:
        dets_ts[det].set_active_time_interval(('%f-%f'%(fluence_start,fluence_stop)))
        dets_spectra[det] = dets_ts[det].to_spectrumlike()
        if (det == "b0" or det == "b1"):
            dets_spectra[det].set_active_measurements(det_b_erange)
        else:
            dets_spectra[det].set_active_measurements(det_n_erange,exclude=['30-35'])


    # plots

    temp_dir = 'unbinned_lightcurves_and_spectra'
    lcdt = 0.5
    if(path_exists_and_is_directory(temp_dir)==False):
        if_directory_not_existing_then_make(temp_dir)
        with within_directory(temp_dir):
            for det in dets:
                fig = dets_ts[det].view_lightcurve(start=bak_pre_start, stop=bak_post_stop,dt=lcdt)
                fig.set_size_inches(10, 8)
                fig.savefig('%s_%s_lc'%(GRB_name,det),bbox_inches="tight", frameon=True)
                with plt.style.context('classic'):
                    fig = dets_spectra[det].view_count_spectrum()
                    fig.set_size_inches(10, 8)
                    fig.savefig('%s_%s_cs'%(GRB_name,det),bbox_inches="tight", frameon=True)
```
